```python
from datasets import load_dataset

dataset = load_dataset("databricks/databricks-dolly-15k")
dataset
```

```python
#  Gemma instruct Chat Template :
#                                 <bos><start_of_turn>user
#                                 Write a hello world program<end_of_turn>
#                                 <start_of_turn>model
```

```python
def generate_prompt(data_point):
    prefix_text = 'Below is an instruction that describes a task. Write a response that appropriately completes the request.\n\n'

    if data_point['context']:
        text = f"""<bos><start_of_turn>user {prefix_text} {data_point["instruction"]}. \n Here are the context: {data_point["context"]} <end_of_turn>\n<start_of_turn>model \n {data_point["response"]} <end_of_turn>"""
    else:
        text = f"""<bos><start_of_turn>user {prefix_text} {data_point["instruction"]} <end_of_turn>\n<start_of_turn>model \n {data_point["response"]} <end_of_turn>"""
    return text

text_column = [generate_prompt(data_point) for data_point in dataset["train"]]
dataset = pd.DataFrame(text_column, columns=["text"])
dataset = Dataset.from_pandas(dataset)
dataset
```
